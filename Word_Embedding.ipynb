#Importing required libraries

```
from nltk.tokenize import sent_tokenize, word_tokenize 
import warnings 
import gensim 
from gensim.models import Word2Vec 
import nltk
nltk.download('punkt')
warnings.filterwarnings(action = 'ignore') 
```

#Reading the text file

http://www.gutenberg.org/files/11/11-0.txt

```
sample = open("/content/11-0.txt", "r") 
s = sample.read()
```

```
print(s)
```

```
# Replaces escape character with space 
f = s.replace("\n", " ")
```

```
print(f)
```

#Data preprocessing

```
data = [] 
  
# iterate through each sentence in the file 
for i in sent_tokenize(f): 
    temp = [] 
      
    # tokenize the sentence into words 
    for j in word_tokenize(i): 
        temp.append(j.lower()) 
  
    data.append(temp) 
```

```
data
```

#CBoW Model

```
# Create CBOW model 
model1 = gensim.models.Word2Vec(data, min_count = 1, size = 100, window = 5) 
```

```
model1
```

```
# Print results 
print("Cosine similarity between 'alice' " + 
               "and 'wonderland' - CBOW : ", 
    model1.similarity('associated', 'agreement')) 
      
print("Cosine similarity between 'alice' " +
                 "and 'machines' - CBOW : ", 
      model1.similarity('agree', 'electronic'))
```

#Skip Gram Model

```
# Create Skip Gram model 
model2 = gensim.models.Word2Vec(data, min_count = 1, size = 100, window = 5, sg = 1)
```

```
model2
```

```
# Print results 
print("Cosine similarity between 'alice' " +
          "and 'wonderland' - Skip Gram : ", 
    model2.similarity('associated', 'agreement')) 
      
print("Cosine similarity between 'alice' " +
            "and 'machines' - Skip Gram : ", 
      model2.similarity('agree', 'electronic'))
```
